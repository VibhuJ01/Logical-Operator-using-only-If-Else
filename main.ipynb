```python
def And(a,b):
    
    if (a == 1):
        if(b == 1):
            return 1

    return 0

def Or(a,b):
    
    if (a == 0):
        if(b == 0):
            return 0

    return 1

def Xor(a,b):

    if(a == b):
        return 0

    return 1

def xNor(a,b):

    if(a != b):
        return 0

    return 1   

def Not(a):

    if(a == 1):
        return 0

    return 1

def main():
    print('\n-------------------------------------------------\n')
    print("1. And Operator")
    print("2. Or Operator")
    print("3. Not Operator")
    print("4. Xor Operator")
    print("5. xNor Operator")
    ch = int(input("What do you want to do? "))

    if(ch != 3):
        a = int(input("\nFirst Input: "))
        b = int(input("Second Input: "))
        if(ch == 1):
            print("\nOutput: ",And(a,b))

        elif(ch == 2):
            print("\nOutput: ",Or(a,b))

        elif(ch == 4):
            print("\nOutput: ",Xor(a,b))

        else:
            print("\nOutput: ",xNor(a,b))

    else:
        a = int(input("\nInput: "))
        print("\nOutput: ",Not(a))

    
    ch1 = input("\nDo you want to continue?(y/n) ")
    if(ch1.lower() == "y"):
        main()

    

    


if(__name__ == "__main__"):
    print('\n-------------------------------------------------\n')
    print("This progam is to showcase logical operator using only  If-Else")
    main()
    print('\n-------------------------------------------------\n')
```
